```python
from matplotlib import (
    pyplot as plt
)
```

## Scatterplot: Brain Volume and IQ

```python
# Brain volume (cm3): 1173 1067 1347 1029 1204
# IQ:                 101 93 94 97 113
```

```python
x = [1173, 1067, 1347, 1029, 1204]
y = [101, 93, 94, 97, 113]

plt.title('Brain Volume and IQ')
plt.xlabel('Brain Volume (cm^3)')
plt.ylabel('IQ', rotation=0)
plt.grid(True)

plt.scatter(x, y)

plt.savefig('../../Homework/02.4/assets/brain-volume-and-iq.png')
plt.show()
```

## Scatterplot: Heights of Fathers and Sons


```python
# Height of father (in.):       73.0 75.5 75.0 75.0 75.0 74.0 74.0 73.0 73.0 78.5
# Height of first son (in.):    74.0 73.5 71.0 70.5 72.0 76.5 74.0 71.0 72.0 73.2
```
