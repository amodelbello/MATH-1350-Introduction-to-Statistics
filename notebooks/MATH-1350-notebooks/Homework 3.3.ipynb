```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../Book Data Sets/01 - Body Data.txt', sep='\t')

males = []
females = []
missing_f_scale_count = 0
for index, row in df.iterrows():
    if (row['GENDER (1=M)'] == 0):
        females.append(row['PULSE'])
    if (row['GENDER (1=M)'] == 1):
        males.append(row['PULSE'])

males.sort()
females.sort()
print(f'males: {males}')
# print(f'females: {females}')

[m_q1, m_q2, m_q3] = np.percentile(males, [25, 50, 75])
[f_q1, f_q2, f_q3] = np.percentile(females, [25, 50, 75])

male_summary = [males[0], m_q1, m_q2, m_q3, males[-1]]
female_summary = [females[0], f_q1, f_q2, f_q3, females[-1]]

# print(male_summary)
# print(female_summary)

plt.title('Pulse Rates')
plt.ylabel('')
plt.xlabel('BPM')
# plt.grid(True)
plt.boxplot(males, positions=[1], labels=['Males'], vert=False)
plt.boxplot(females, positions=[2], labels=['Females'], vert=False)
plt.savefig('../../Homework/03.3/assets/pulse-rates.png')
plt.show()
```
